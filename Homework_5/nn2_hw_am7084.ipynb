# PROBLEM 1 (40%)

In the lecture notebook we discussed how a neural network autoencoder trained with respect to the mean squared error objective function and no non-linearity can reproduce PCA. In this problem we want to show this explicitly. First, take the 0 and 1 digits from the sklearn digits dataset. 

```python
from __future__ import print_function, division

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns; sns.set()
import pandas as pd

from sklearn.datasets import load_digits
digits = load_digits()
selector = digits.target<2
X = digits.data[selector]
Y = digits.target[selector]
```

Next, apply PCA to the dataset (ignoring the digit labels) and plot the resulting dimensional reduction down to $d=2$ (i.e. the projection onto the two leading principal components). Label each datapoint according to its digit label (e.g. red and blue). So far this is just reproducing what we've already done in the lecture notebook.

```python
from sklearn.decomposition import PCA
pca = PCA(2)  # project from 64 to 2 dimensions
Xproj = pca.fit_transform(X)
plt.scatter(Xproj[:, 0], Xproj[:, 1], c=Y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar();
```

But now we want to reproduce the same plot (or at least a very similar plot) using a neural network autoencoder. To do this you should do the following steps:

- use the FunctionApproximator class to train a neural network with the output vectors ($y_i$) equal to the correspoding input vectors ($x_i$), i.e. $y_i = x_i$ for all $i$ 


- use only two hidden units (this way we'll only get the two leading principal components) and set the activation function (i.e. the non-linearity) to none ('activation=None' as an argument to FunctionApproximator); note that the cost function is the mean squared error by default so we don't need to specify that explicitly


- use the get_y_pred() function in FunctionApproximator to get the vectors outputted by the neural network on the input vectors; note that although these predicted vectors are 64 dimensional they all in fact live in a 2-dimensional subspace, since they all had to pass through the 2 hidden units


- run PCA on these predicted output vectors and plot the transformed dataset by projecting onto the two leading principal components (basically this will just find the relevant two-dimensional subspace within the set of output vectors; the 'actual' PCA was done by the autoencoder it's just that we use the PCA class to easily identify the subspace that the autoencoder has chosen)


- plot the 'classical' PCA result next to the 'autoencoder' PCA result and comment briefly on the degree to which they are the same (make sure you've trained the neural network reasonably well; and note that the results may differ by a sign so you may want to transform the autoencoder result by a sign change to make things more directly comparable)

```python
y = X
```

```python
from function_approximator import FunctionApproximator
fa = FunctionApproximator(n_out=64,n_in=64, n_hidden=2, activation=None)
fa.train(X,y, learning_rate=0.005, n_epochs=200000, report_frequency=20000)
Y_pred = fa.get_y_pred()

```

```python
Y_pred
```

```python
Y_pred.shape
```

```python
Xproj2 = pca.fit_transform(Y_pred)
```

```python
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)
fig = plt.figure(figsize=[18,8]);
ax1 = plt.subplot(121);
plt.scatter(Xproj2[:, 0], Xproj2[:, 1], c=Y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('GnBu', 10));
ax1.set_title('faux autoencoder',**font);
plt.colorbar();
ax2 = plt.subplot(122);
plt.scatter(Xproj[:, 0], Xproj[:, 1], c=Y, edgecolor='none', alpha=0.5,
            cmap=plt.cm.get_cmap('nipy_spectral', 10));
ax2.set_title('pca', **font);
plt.colorbar();
```

They look extremely similar, the shape is roughly the same and there is really no overlap. They both do a good job seperating the digits. However, PCA looks like it does a better job of classifying the digits.

```python
deaths = pd.read_csv('deaths-from-homicides.csv')
deaths.head()
```

#### (A)

```python
#time series of firearm vs non-firearm homicides in australia from 1915 until 2004
#per 100000 people
fig = plt.figure(figsize=[10,8])
plt.plot(deaths['Year'],deaths['Firearm'], label='Firearm')
plt.plot(deaths['Year'],deaths['Non firearm'], label='Non firearm')
plt.legend(loc='upper left',prop={'size':18})
plt.ylabel('Deaths per 100,000')
plt.xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

#### (B)

```python
deaths['estimate_NF'] = (deaths['Non firearm'].mean()/deaths['Firearm'].mean())*deaths['Firearm']
deaths.head()
```

```python
fig = plt.figure(figsize=[10,8])
plt.plot(deaths['Year'],deaths['estimate_NF'], label='Esimate Non firearm')
plt.plot(deaths['Year'],deaths['Non firearm'], label='Non firearm')
plt.legend(loc='upper left',prop={'size':12})
plt.ylabel('Deaths per 100,000')
plt.xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

#### (C)

```python
from rnn import RNN, MetaRNN
```

```python
#n_epochs = 200000
n_hidden = 3
n_in = 1
n_out = 1
n_steps = 90
n_seq = 1

seq = deaths['Firearm'].reshape((n_seq,n_steps,n_in))
targets = deaths['Non firearm'].reshape((n_seq, n_steps, n_out))

```

```python
model = MetaRNN(n_in=n_in, n_hidden=n_hidden, n_out=n_out,
                    learning_rate=0.1, n_epochs=50000,
                activation='tanh', learning_rate_decay=0.999)
model.fit(seq, targets, validation_frequency=10000)
guess = model.predict(seq[0])
```

```python
fig = plt.figure(figsize=[10,8])
plt.plot(deaths['Year'],deaths['estimate_NF'], label='Esimate Non firearm')
plt.plot(deaths['Year'],deaths['Non firearm'], label='Non firearm')
plt.plot(deaths['Year'],guess, marker='*',markersize=8,linestyle='---', label='RNN Estimate')
plt.legend(loc='upper left',prop={'size':12})
plt.ylabel('Deaths per 100,000')
plt.xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

The RNN seems to be more constant than the estimated non firearm deaths. Visually the RNN curve does not dip and rise as much as the other two, but it appears to follow similar trends as the other two do.
